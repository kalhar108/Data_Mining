```python
import pandas as pd
from pycaret.clustering import *
```

```python
data = pd.read_csv('~/Wholesale customers data.csv')

data.head()
```

```python
data.info()
```

```python
cluster_setup = setup(data, normalize=True, ignore_features=['Channel', 'Region'], use_gpu=True)
```

```python
k_means = create_model('kmeans')

clustered_data = assign_model(k_means)

clustered_data.head()
```

```python
plot_model(k_means, plot='cluster')
```
