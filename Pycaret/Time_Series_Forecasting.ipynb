```python
import pandas as pd
from pycaret.time_series import *
```

```python
data = pd.read_csv('~/DailyTemp.csv', parse_dates=['Date'], index_col='Date')
data = data.asfreq('D')
data['Temp'] = pd.to_numeric(data['Temp'], errors='coerce')
data['Temp'] = data['Temp'].fillna(method='ffill')
print(data.head())
```

```python
ts_setup = setup(
    data=data,
    target='Temp',
    session_id=123,
    seasonal_period=365,
    fold=5,
    use_gpu=True
)
```

```python
arima_model = create_model('arima')
```

```python
best_model = compare_models()
```

```python
final_model = finalize_model(best_model)
```

```python
future_forecast = predict_model(final_model, fh=30)

print(future_forecast)
```

```python
plot_model(final_model, plot='forecast', data_kwargs={'fh': 30})
```
